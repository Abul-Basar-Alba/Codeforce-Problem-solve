```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Define the function f(x) and its rearrangement g(x)
def f(x):
    return x**3 - 2*x - 2

def g(x):
    # Rearranged form: x = (2x + 2)^(1/3)
    return (2*x + 2)**(1/3)

```

```python
# Define the function and its rearrangement for iteration
def iteration_method(g, x, tol=1e-5, max_iter=100):
    for i in range(max_iter):
        x_new = g(x)
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    raise ValueError(
        "Iteration did not converge within the maximum number of iterations."
    )
root = iteration_method(g, -1)
print(f"Root found: {root:.6f}")

```

```python
# Let's plot the result
x = np.arange(-10, 10, 0.1)

plt.plot(x, f(x))

plt.scatter(root, f(root), color='blue')  # Mark the root on the plot
plt.axvline(root, color='purple', linestyle='--')

plt.axvline(0, color='green', linestyle='--')
plt.axhline(0, color='red', linestyle='--')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f"Iteration Method Root: {root:.5f}")
plt.grid()
plt.show()
```
