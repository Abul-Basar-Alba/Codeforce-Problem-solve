```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2-2

def bisection_method(func,a,b,tol=1e-5):
    if func(a)*func(b)>=0:
        raise ValueError("Function values at the endpoints must have opposite signs.")
    mid=(a+b)/2.0
    if abs(func(mid))<tol:
       return mid
    elif func(a)*func(mid)<0:
        return bisection_method(func,a,mid,tol)
    else:
        return bisection_method(func,mid,b,tol)

root=bisection_method(f,0,10)

x=np.arange(0,3,0.1)
print(x)

plt.plot(x,f(x))
plt.axvline(root)

plt.grid()
plt.show()
```
